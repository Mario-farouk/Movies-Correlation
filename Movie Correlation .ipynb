```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8)




data=pd.read_csv("D:\GitHup\Python Projects\Movies\movies.csv")
data
```

```python
data.isnull().sum()
```

```python
budg_mean=data['budget'].mean()
grss_mean=data['gross'].mean()
runtime_mean=data['runtime'].mean()
votes_mean=data['votes'].mean()
```

```python
data['budget']=data['budget'].fillna(budg_mean)
data['gross']=data['gross'].fillna(grss_mean)
data['rating']=data['rating'].fillna('R')
data['company']=data['company'].fillna('Universal Pictures')
data['votes']=data['votes'].fillna(0)
data['runtime']=data['runtime'].fillna(0)
data['votes']=data['votes'].replace(0,votes_mean)
data['runtime']=data['runtime'].replace(0,runtime_mean)
```

```python
# DataType Cleaning
data.info()
```

```python
data['budget']=data['budget'].astype('int64')
data['gross']=data['gross'].astype('int64')
data['votes']=data['votes'].astype('int64')
data['runtime']=data['runtime'].astype('int64')
```

```python
data.info()
```

```python
# Now data is cleaned we can apply our analysis
data
```

```python
data.describe()
```

```python
# lets start a correlation methods 
# budget with gross 
# company 
```

```python
# scatter plot with budget vs gross 
plt.scatter(x=data['budget'],y=data['gross'])

plt.title('Budget vs Gross')
plt.xlabel('gross earning')
plt.ylabel('budget spent')

plt.show()
```

```python
# plot budget with seaborn
sns.regplot(x='budget' ,y='gross' ,data=data ,scatter_kws={'color':'red'},line_kws={'color':'blue'})
```

```python
data.corr(method ='spearman') #pearson , kendall , spearman
```

```python
correlation_method = data.corr(method ='spearman')
sns.heatmap(correlation_method,annot=True)
plt.title('Correlation Matrix')
plt.xlabel('film matrix')
plt.ylabel('film matrix')
plt.show()
```

```python
# let's numerized data for more correlation 
```

```python
data_numoraized=data
for col_name in data_numoraized.columns:
    if(data_numoraized[col_name].dtype == 'object'):
        data_numoraized[col_name] = data_numoraized[col_name].astype('category')
        data_numoraized[col_name] = data_numoraized[col_name].cat.codes

data_numoraized
```

```python
correlation_method = data_numoraized.corr(method ='spearman')
sns.heatmap(correlation_method,annot=True)
plt.title('Correlation Matrix')
plt.xlabel('film matrix')
plt.ylabel('film matrix')
plt.show()
```
